Task 2: Data Preprocessing and Feature Engineering

```python
# Check missing values
df.isnull().sum()

# Fill numerical missing values with median
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical missing values with mode
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

# Remove outliers using IQR method
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR)))]
```

```python
from sklearn.feature_selection import mutual_info_regression

X = df.drop('VisitorCount', axis=1)
y = df['VisitorCount']

# Encode non-numeric features temporarily for feature scoring
X_encoded = pd.get_dummies(X, drop_first=True)

# Mutual Info
mi_scores = mutual_info_regression(X_encoded, y)
mi_df = pd.DataFrame({'Feature': X_encoded.columns, 'Score': mi_scores})
top_features = mi_df.sort_values(by='Score', ascending=False).head(5)
print(top_features)
```
